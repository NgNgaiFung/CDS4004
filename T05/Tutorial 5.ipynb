### Part 1 - Design our first graph

``networkx`` provides support for several graph models.

Among them:

undirected graphs, available through the Graph class
directed graphs, available through the DiGraph class
In this brief tutorial we will focus only on undirected graphs.

#### We can build a graph adding nodes as follows:

```python
g = nx.Graph()
g.add_node('a')
g.add_node('b')
g.add_node('c')

fig = plt.figure(figsize=(2,2))
nx.draw(g ,with_labels=True)
```

#### We can build a graph adding edges as follows:

```python
g1 = nx.Graph()
g1.add_edge('a', 'b')
nx.draw(g1 ,with_labels=True)
```

#### We can also add edges directly without adding the nodes

```python
g1 = nx.Graph()
g1.add_edge('a', 'b')
g1.add_edge('a', 'c')
g1.add_edge('a', 'd')

fig = plt.figure(figsize=(2,2))
nx.draw(g1 ,with_labels=True)
```

#### Nodes and edges can be removed

```python
g1.remove_node("c")
g1.remove_edge("a", "d")

#re-draw the graph
fig = plt.figure(figsize=(2,2))
nx.draw(g1, with_labels=True)
```

#### Another way to add edges: add_edges_from()

```python
g2 = nx.Graph()
g2.add_edges_from(g1.edges)
#Note that: node d in g1 is ignored due to no edges
g2.add_edge('b', 'NN')
fig = plt.figure(figsize=(2,2))
nx.draw(g2 ,with_labels=True)
```

```python
g3 = nx.Graph()
# Note that: nodes will be added only, edges are not added
g3.add_nodes_from(g2)
fig = plt.figure(figsize=(2,2))
nx.draw(g3 ,with_labels=True)
```

#### Task 1.2 Create a network g9 with 9 nodes ranging from 0-8 and form a circle (circle graph)

```python
#Please write your code here

n = 9
g9 = nx.Graph()
for i in range(n):
    g9.add_node(i)
for i in range(n):
    g9.add_edge(i, (i+1)%n)
fig = plt.figure(figsize=(2,2))
nx.draw(g9 ,with_labels=True)
```

#### Task 1.3 Create a network g10 with 10 nodes ranging from 0-9 and connect them with each other (complete graph)

```python
#Please write your code here

n = 10
g10 = nx.Graph()

for i in range(n):
    g10.add_node(i)
for i in range(n):
    for j in range(n):
        if i != j:
            g10.add_edge(i, j)
fig = plt.figure(figsize=(2,2))
nx.draw(g10 ,with_labels=True)
```

#### Task 1.4 Create a network g11 with 11 nodes connect in a line

### Part 3 - Reading a graph from file

``networkx`` natively supports several network file formats.

Among them one the most frequently used in online repository is the *edgelist* one.

An edge list is a text file (usually saved as .csv) in which each line identifies an edge. <br/>
For instance, the triangle defined before can be described as:

    a,b
    b,c
    c,a

#### Execute the cell below to read edgelist file:

```python
g = nx.read_edgelist("network_demo.csv", delimiter=",", nodetype=int)
```

```python
nx.draw(g ,with_labels=True)
```

Similarly a graph can be written to file using ``nx.write_edgelist(g, filename)``.

For all the I/O methods refer to the [official documentation](https://networkx.github.io/documentation/latest/reference/readwrite/index.html)

#### Task 3.1 Remove odd nodes and save to a file "even_nodes.csv"

```python
#Please write your code here
arr = []
for i in g.nodes():
    if i%2 != 0:
        arr.append(i)

for i in arr:
    g.remove_node(i)

fig = plt.figure(figsize=(2,2))
nx.draw(g ,with_labels=True)

print(g.edges())

nx.write_edgelist(g, "even_nodes.csv")
```
